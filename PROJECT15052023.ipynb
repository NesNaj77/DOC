```python
#KMEANS#
import pandas as pd
from sklearn.cluster import KMeans
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances=[]
K=range(1,10)
for k in K:
    kmeans=KMeans(n_clusters=k,random_state=0)
    kmeans.fit(data1)
    Sum_of_squared_distances.append(kmeans.inertia_)
    labels=pd.DataFrame(kmeans.labels_)
    labels
    kmeans.predict(data1)
    print(kmeans.cluster_centers_)
```

```python
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('ElbowMethod')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=0)
kmeans.fit(data1)
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
#KNN#
#Choosing the features as x and the target as y
x=data1[['Corporate','Direct','GDS','TA/TO','Undefined']]
y=data1['reservation_status']
data1


#import the relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30) #split our data with test size of 20% 

knn=KNeighborsClassifier(n_neighbors=20) #build our knn classifier
knn.fit(x_train,y_train) #Training KNN classifier
y_pred=knn.predict(x_test)  #Testing
print('Acuuracy=',accuracy_score(y_pred,y_test))
```

```python
n_neighbors=20
scores=[]
for k in range(1,20):
    knn=KNeighborsClassifier(n_neighbors-k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    print('Accuracy for k=',k,'is:',round(accuracy_score(y_pred,y_test),2))
    scores.append(round(accuracy_score(y_pred,y_test),2))
```

```python
plt.plot(range(1,20),scores)
plt.xlabel('Value K for KNN')
plt.ylabel('Testing Accuracy')
```
