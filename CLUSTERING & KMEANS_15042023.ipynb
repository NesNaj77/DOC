```python
import pandas as pd
data=pd.read_csv("CCGENERAL.csv")
data
```

```python
data=data.drop("CUST_ID",axis=1)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data["MINIMUM_PAYMENTS"].value_counts()
```

```python
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data["CREDIT_LIMIT"].value_counts()
```

```python
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='complete')
clust_labels=model.fit_predict(data)
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
ax=fig.add_subplot(111)
scatter=ax.scatter(data['PURCHASES'],data['CASH_ADVANCE'],c=agglomerative[0],s=50)
ax.set_title("Agglomerative Clustering")
ax.set_xlabel("PURCHASES_FREQUENCY")
ax.set_ylabel("CREDIT_LIMIT")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data,method="complete"))
```

```python
from sklearn.cluster import KMeans
Sum_of_squared_distances=[]
K=range(1,10)
for k in K:
    kmeans=KMeans(n_clusters=k,random_state=0)
    kmeans.fit(data)
    Sum_of_squared_distances.append(kmeans.inertia_)
    labels=pd.DataFrame(kmeans.labels_)
    labels
    kmeans.predict(data)
    print(kmeans.cluster_centers_)
```

```python
plt.plot(K,Sum_of_squared_distances,'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('ElbowMethod')
plt.show()
```

```python
kmeans=KMeans(n_clusters=4,random_state=0)
kmeans.fit(data)
labels=pd.DataFrame(kmeans.labels_)
labels
```
